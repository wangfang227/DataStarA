```python
wordcount={}
songs=song.split()
```

```python
for word in songs:
    word=word.encode('utf-8','ignore')
    word=word.lower()
    #wordlist.append(word)
    if word in wordcount.keys():
        wordcount[word]=wordcount[word]+1
    else:
        wordcount[word]=1
    
totalsum=sum(wordcount.values())
wordfreq={key:round(float(value)/totalsum,6) for key, value in wordcount.items()}
wordcount
wordfreq
            
topwords=sorted(wordcount.items(),key=lambda x:x[1],reverse=True)
```

```python
keywords=dict((key,value) for (key,value) in wordcount.items() if value>=10)
keywords=sorted(keywords.items(),key=lambda x:x[1],reverse=True)
```

```python
keywords_str=' '.join([x[0].decode("utf-8")  for x in keywords])
keywords_str
```

```python
stopwords=['the','to','is','are','am','of','in','a','if','have',
           'can','could','that','this','do','does','did','be','as',
           'will','(female)','(m)','(male)','~','fang','vincent','jay','music','and','on','for','too','so','even','but',
          'just','at','was','had']
song=' '.join([x.lower() for x in song.split() if x.lower() not in stopwords])
songs=song.split()
songdict0=nltk.FreqDist(songs)
songdict=sorted(songdict0.items(),key=lambda x:x[1],reverse=True)
topbidict=songdict[:100]
topbidict
songdict10=[x for x in songdict if x[1]>=10]
songdict10
#songdict0.most_common()
#songdict0.max()
#songdict0.plot()
```

```python
bigrams=list(nltk.bigrams(songs))
bidict0=nltk.FreqDist(bigrams)
bidict=sorted(bidict0.items(),key=lambda x:x[1],reverse=True)
topbidict=bidict[:100]
topbidict
bidict10=[x for x in bidict if x[1]>=10]
bidict10
#sbidict0.most_common()
#bidict0.max()
#bidict0.plot()
```

```python
from nltk.util import ngrams
ngrams=list(nltk.ngrams(songs,3))
ngrams
ndict0=nltk.FreqDist(ngrams)
ndict=sorted(ndict0.items(),key=lambda x:x[1],reverse=True)
topndict=ndict[:100]
topndict
ndict10=[x for x in ndict if x[1]>=3]
ndict10
```
